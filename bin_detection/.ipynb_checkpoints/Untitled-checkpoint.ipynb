```python
#Better results with my data?
X1 = np.load("color_training_data/test-bin_blue.npy").astype(np.float64)/255
X2 = np.load("color_training_data/test-other_colors.npy").astype(np.float64)/255
X3 = np.load("color_training_data/red.npy").astype(np.float64)/255
X4 = np.load("color_training_data/green-original.npy").astype(np.float64)/255
X5 = np.load("color_training_data/black.npy").astype(np.float64)/255
```

```python
X1 = np.delete(X1, 0, 0)
X2 = np.delete(X2, 0, 0)
X3 = np.delete(X3, 0, 0)
X4 = np.delete(X4, 0, 0)
X5 = np.delete(X5, 0, 0)
print(X1.shape)
print(X2.shape)
print(X3.shape)
print(X4.shape)
print(X5.shape)
```

```python
y1, y2, y3, y4, y5 = np.full(X1.shape[0],1), np.full(X2.shape[0], 2), np.full(X3.shape[0],3), np.full(X4.shape[0],4),np.full(X5.shape[0],5)
```

```python
X, y = np.concatenate((X1,X2,X3, X4, X5)), np.concatenate((y1,y2,y3, y4, y5))
```

```python
y[104942]
```

```python
from pixel_classifier import PixelClassifier
```

```python
multilogistic = PixelClassifier()
weigths = multilogistic.train(X, y, lr=0.0001, verbose=False)
print(multilogistic.score(X, y))
```

```python
fig = plt.figure(figsize=(8,6))
plt.plot(np.arange(len(multilogistic.loss)), multilogistic.loss)
plt.title("Development of loss during training")
plt.xlabel("Number of iterations")
plt.ylabel("Loss")
plt.show()
```

```python
from bin_detector import BinDetector
import numpy as np
import matplotlib.pyplot as plt; plt.ion()
import os, cv2
```

```python
my_detector = BinDetector()
img = cv2.imread("data/validation/0067.jpg")
#img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
rows, cols, channels = img.shape
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
img_ = np.copy(img)
mask_img = my_detector.segment_image(img_)
```

```python

estm_boxes = my_detector.get_bounding_boxes(mask_img)
plt.imshow(mask_img, cmap='gray')
print(estm_boxes)
```

```python
image =cv2.rectangle(img, (estm_boxes[1][0], estm_boxes[1][1]), (estm_boxes[1][2]
                                                                                           , estm_boxes[1][3]), (255,0,0), 5)
plt.imshow(image)
```
